```python
import pandas as pd
import numpy as np
import ast
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
stopwords = set(stopwords.words('english'))
stemmer = PorterStemmer()
```

```python
def tokenize(text):
    # make lowercase and tokenize
    tokens = word_tokenize(text)
    tokens = [token.lower() for token in tokens]
    return tokens
```

```python
def removeStopwords(tokens):
    tokens = [token for token in tokens if token not in stopwords]
    
    return tokens
```

```python
def regex(tokens):
    tokens = [re.sub(r'[^\w\s]','',token) for token in tokens]

    tokens = [token for token in tokens if token.isnumeric() != True and token != '']
    
    return tokens
```

```python
def stemming(tokens):
    tokens = [stemmer.stem(token) for token in tokens]
    
    return tokens
```

```python
def cleanText(text):
    
    tokens = tokenize(text)

    # remove stopwords
    tokens = removeStopwords(tokens)
    
    tokens = regex(tokens)
    
    cleaned = stemming(tokens)
    

    return cleaned
```

```python
def findAnxietyWordMatches(clean_tokens):
    aw=[]
    [aw.append(i) for i in clean_tokens if i in anxiety_words]

    return aw
```

```python
def findNegWordMatches(clean_tokens):
    nw=[]
    [nw.append(i) for i in clean_tokens if i in neg_words]

    return nw
```

```python
def findPosWordMatches(clean_tokens):
    pw=[]
    [pw.append(i) for i in clean_tokens if i in pos_words]

    return pw
```

```python
def getTotalWords(comments):
    total_wc = 0

    comment_wc=[len(comment.split()) for comment in comments]
    
    for wc in comment_wc:
        total_wc += wc

    return total_wc
```

```python
def getDataFrame(company):
    df = pd.read_csv(f'./ARTICLES/{company}/{company}_articles.csv',
                     converters={'Comments':ast.literal_eval},
                     usecols=['Date','Headline','Comments'],
                     index_col='Date')
    df['Comments']=df['Comments'].apply(lambda article: [comment for comments in article for comment in comments])
    df['Cleaned Tokens']=df['Comments'].apply(lambda article:[cleanText(comment) for comment in article]).apply(lambda article:[comment for comments in article for comment in comments])
    df['Total Word Count']=df['Comments'].apply(lambda x: getTotalWords(x))
    df['Anxiety Words Found']=df['Cleaned Tokens'].apply(lambda x: findAnxietyWordMatches(x))
    df['Total Anxiety Words']=df['Anxiety Words Found'].apply(lambda x: len(x))
    df['Anxiety Score']=(df['Total Anxiety Words']/df['Total Word Count']*100).round(decimals=2)
    
    df['Negative Words Found']=df['Cleaned Tokens'].apply(lambda x: findNegWordMatches(x))
    df['Total Negative Words']=df['Negative Words Found'].apply(lambda x: len(x))
    df['Negativity Score']=(df['Total Negative Words']/df['Total Word Count']*100).round(decimals=2)
    
    df['Positive Words Found']=df['Cleaned Tokens'].apply(lambda x: findPosWordMatches(x))
    df['Total Positive Words']=df['Positive Words Found'].apply(lambda x: len(x))
    df['Positivity Score']=(df['Total Positive Words']/df['Total Word Count']*100).round(decimals=2)
    
    df.drop('Cleaned Tokens',1,inplace=True)
    
    df.to_csv(f'./TextAnalysis/{company}_anx.csv',index=True,mode='w+')
    
    return df
```
